### House Prices Prediction

In this exercise you'll try to build a neural network that predicts the price of a house according to a simple formula.

So, imagine if house pricing was as easy as a house costs 50k + 50k per bedroom, so that a 1 bedroom house costs 100k, a 2 bedroom house costs 150k etc.

How would you create a neural network that learns this relationship so that it would predict a 7 bedroom house as costing close to 400k etc.

Hint: Your network might work better if you scale the house price down.
You don't have to give the answer 400...it might be better to create something that predicts the number 4, and then your answer is in the 'hundreds of thousands' etc.

```python
hc = 50000
pbr = 50000
def get_house_price(no_of_bedrooms):
    return hc+pbr*no_of_bedrooms
```

```python
get_house_price(4)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
import random
```

```python
xs = [random.randrange(1,10) for i in range(100)]
ys = [get_house_price(x) for x in xs]
xs = np.array(xs)
ys=np.array(ys)
# scaling ys
ys = ys/100000
```

```python
xs[:10]
```

```python
ys[:10]
```

```python
model = keras.models.Sequential([keras.layers.Dense(units=1,input_shape = [1])])
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
model.fit(xs,ys,epochs=500)
```

```python
model.predict([4])
```
